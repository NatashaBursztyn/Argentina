### Variables:	
Gdp: Gross domestic product per dollar, PBI de cada provincia

illiteracy: tasa de población no educada, en %

poverty: tasa de pobreza por provincia, en %	

deficient_infra: falta de infraestructura	

school_dropout:	tasa de gente que deja la escuela

no_healthcare: % de población sin acceso a servicios de salud	

birth_mortal: tasa de mortalidad al nacimiento	

pop: población	

movie_theatres_per_cap: tasa de cine per cápita	

doctors_per_cap: doctores disponibles per cápita
